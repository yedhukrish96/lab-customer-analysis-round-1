```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv('file1.csv')
file1
```

```python
file2= pd.read_csv('file2.csv')
file2
```

```python
file3= pd.read_csv('file3.csv')
file3
```

```python
#standardising header names
cols1= []
for c in file1.columns:
    cols1.append(c.lower())
file1.columns = cols1
file1
```

```python
cols2 = []
for c in file2.columns:
    cols2.append(c.lower())
file2.columns = cols2
file2
```

```python
cols3 = []
for c in file3.columns:
    cols3.append(c.lower())
file3.columns = cols3
file3
```

```python
file1 = file1.rename(columns = {'st':'state', })
file2 = file2.rename(columns = {'st': 'state'})
file1 = file1[['customer', 'state', 'gender', 'education', 'customer lifetime value','income','monthly premium auto','number of open complaints','policy type','vehicle class','total claim amount']]
file2= file2[['customer', 'state', 'gender', 'education', 'customer lifetime value','income','monthly premium auto','number of open complaints','policy type','vehicle class','total claim amount']]
file3= file3[['customer', 'state', 'gender', 'education', 'customer lifetime value','income','monthly premium auto','number of open complaints','policy type','vehicle class','total claim amount']]
file3
```

```python
#concatenating the three data frames
data= pd.concat([file1,file2,file3],axis=0)
data

```

```python
#which columns are numerical?
data.dtypes
```

```python
num_cols = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

num_cols = data.select_dtypes(include=num_cols)
num_cols
```

```python
#which columns are catagorical?
cat_cols= ['object']
cat_cols = data.select_dtypes(include=cat_cols)
cat_cols
```

```python
#deleting the colum education and number of open complaints
data= data.drop(['education', 'number of open complaints'], axis=1)
data
```

```python
#converting percentages to numerical values
data['customer lifetime value'] = data['customer lifetime value'].str.rstrip('%').astype('float') / 100.0 #all thanks to stackoverflow.com
data
```

```python
#checking for duplicate values
duplicates= data.duplicated()
duplicates.sum()
```

```python
#deleting the duplicate values
data= data.drop_duplicates()
data
```

```python
#customers who have income less than 0
lowinc= data[data['income']<=0]
lowinc
```
